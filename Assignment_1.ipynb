## In-place Quick Sort

```python
import random
import time

def partition(arr, low, high, comparisons, swaps):
    pivot = arr[high]
    i = low - 1

    for j in range(low, high):
        comparisons += 1
        if arr[j] <= pivot:
            i += 1
            swaps += 1
            arr[i], arr[j] = arr[j], arr[i]

    swaps += 1
    arr[i + 1], arr[high] = arr[high], arr[i + 1]
    return i + 1, comparisons, swaps

def quick_sort(arr, low, high, comparisons, swaps):
    if low < high:
        pivot_index, comparisons, swaps = partition(arr, low, high, comparisons, swaps)
        comparisons, swaps = quick_sort(arr, low, pivot_index - 1, comparisons, swaps)
        comparisons, swaps = quick_sort(arr, pivot_index + 1, high, comparisons, swaps)

    return comparisons, swaps

num_arrays = 100
min_array_size = 1
max_array_size = 20
min_element_value = 0
max_element_value = 100

total_comparisons = 0
total_swaps = 0
st=time.time()

for i in range(num_arrays):
    arr = [random.randint(min_element_value, max_element_value) for _ in range(random.randint(min_array_size, max_array_size))]
    comparisons, swaps = quick_sort(arr, 0, len(arr) - 1, 0, 0)
    total_comparisons += comparisons
    total_swaps += swaps

    print(f"Array",i+1,":",arr)
    print(f"Comparisons: {comparisons}")
    print(f"Swaps: {swaps}\n")

end=time.time()
total_time=end-st

average_comparisons = total_comparisons / num_arrays
average_swaps = total_swaps / num_arrays

print(f"Average Comparisons: {average_comparisons}")
print(f"Average Swaps: {average_swaps}")
print("Total Time",total_time*1000,"milliseconds")
```

## In-place Heap Sort

```python
def heapify(arr, n, i, comparisons, swaps):
    largest = i
    left = 2 * i + 1
    right = 2 * i + 2

    if left < n and arr[left] > arr[largest]:
        comparisons += 1
        largest = left

    if right < n and arr[right] > arr[largest]:
        comparisons += 1
        largest = right

    if largest != i:
        swaps += 1
        arr[i], arr[largest] = arr[largest], arr[i]
        comparisons += 1

        swaps, comparisons = heapify(arr, n, largest, comparisons, swaps)

    return swaps, comparisons

def heap_sort(arr):
    n = len(arr)
    comparisons = 0
    swaps = 0

    for i in range(n // 2 - 1, -1, -1):
        swaps, comparisons = heapify(arr, n, i, comparisons, swaps)

    for i in range(n - 1, 0, -1):
        swaps += 1
        arr[i], arr[0] = arr[0], arr[i]
        comparisons += 1

        swaps, comparisons = heapify(arr, i, 0, comparisons, swaps)

    return swaps, comparisons

num_arrays = 100
min_array_size = 1
max_array_size = 20
min_element_value = 0
max_element_value = 100

total_comparisons = 0
total_swaps = 0
st=time.time()
for i in range(num_arrays):
    arr = [random.randint(min_element_value, max_element_value) for _ in range(random.randint(min_array_size, max_array_size))]
    swaps, comparisons = heap_sort(arr)
    total_comparisons += comparisons
    total_swaps += swaps

    print(f"Array",i+1,":",arr)
    print(f"Comparisons: {comparisons}")
    print(f"Swaps: {swaps}\n")

end=time.time()
total_time=end-st
average_comparisons = total_comparisons / num_arrays
average_swaps = total_swaps / num_arrays

print(f"Average Comparisons: {average_comparisons}")
print(f"Average Swaps: {average_swaps}")
print("Total Time",total_time*1000,"milliseconds")
```

## Merge Sort


## Radix Sort

```python
def countsort(arr,min_element,max_element,exp):
    arr_range = max_element - min_element + 1
    freq = [0]*arr_range
    
    for i in range(len(arr)):
        idx = int(arr[i]/exp)%10 - min_element
        freq[idx]+=1
        
    ans = [0]*len(arr)
    
    for i in range(1,len(freq)):
        freq[i] = freq[i] + freq[i-1]
        
    for i in range(len(arr)-1,-1,-1):
        pos = freq[int(arr[i]/exp)%10 - min_element]-1
        ans[pos] = arr[i]
        freq[int(arr[i]/exp)%10 - min_element]-=1
    
    for i in range(len(arr)):
        arr[i]=ans[i]
```

```python
from collections import deque
from collections import defaultdict

class SnakeAndLaddersGame:
    def __init__(self, n, snakes, ladders):
        self.n = n
        self.snakes = snakes
        self.ladders = ladders
        self.graph = defaultdict(list)

    def construct_board(self):
        for i in range(1, self.n * self.n + 1):
            if i in self.snakes:
                self.graph[i].append(self.snakes[i])
            elif i in self.ladders:
                self.graph[i].append(self.ladders[i])
            else:
                for j in range(1, 7):
                    if i + j <= self.n * self.n:
                        self.graph[i].append(i + j)
                        
    def shortest_sequence(self):
        dist = [-1] * (self.n * self.n + 1)
        dist[1] = 0
        queue = deque([(1, [])]) 

        while queue:
            node, visited_nodes = queue.popleft()
            for neighbor in self.graph[node]:
                if neighbor == self.n * self.n:
                    visited_nodes.append(neighbor)
                    return dist[node] + 1, visited_nodes
                if dist[neighbor] == -1:
                    dist[neighbor] = dist[node] + 1
                    new_visited_nodes = visited_nodes.copy()
                    new_visited_nodes.append(neighbor)
                    queue.append((neighbor, new_visited_nodes))

        return -1, []  # Goal cannot be reached

     

    def checkcycles(self):
        
        snake_tails = set(self.snakes.values())
        ladder_bottoms = set(self.ladders.keys())
        snake_heads = set(self.snakes.keys())
        ladder_heads = set(self.ladders.values())
        #return bool(snake_tails.intersection(ladder_bottoms))&bool(snake_heads.intersection(ladder_heads))
        intersection = snake_tails.intersection(ladder_bottoms)
        intersection1= snake_heads.intersection(ladder_heads)
        
        if intersection:
            print("Cycles found at points:", intersection,"to",intersection1)
            return True
        else:
            return False

    def directladder(self):
        
        for keys, values in self.ladders.items():
            if keys == 1 and values == self.n * self.n:
                return True
        return False
    
    def is_valid_snakes_and_ladders(self):
        positions = set(self.snakes.keys()) | set(self.ladders.keys())
        return len(positions) == len(self.snakes) + len(self.ladders)
```

```python
class KnapsackBacktracking:
    def __init__(self, weights, values, capacity):
        self.weights = weights
        self.values = values
        self.capacity = capacity
        self.n = len(weights)
        self.max_value = 0

    def knapsack(self, i, current_weight, current_value):
        if i == self.n:
            if current_weight <= self.capacity:
                self.max_value = max(self.max_value, current_value)
            return
        
        # Exclude item i
        self.knapsack(i + 1, current_weight, current_value)
        
        # Include item i
        if current_weight + self.weights[i] <= self.capacity:
            self.knapsack(i + 1, current_weight + self.weights[i], current_value + self.values[i])

    def solve(self):
        self.knapsack(0, 0, 0)
        return self.max_value

# Example usage
weights = [10, 20, 30]
values = [60, 100, 120]
capacity = 50
knapsack = KnapsackBacktracking(weights, values, capacity)
print("Maximum value using backtracking:", knapsack.solve())
```
